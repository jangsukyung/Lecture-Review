```python
df = pd.DataFrame({
	"가로" : [10, 20, 30, 10, 30, 20, 11],
	"세로" : [20, 23, 22, 33, 22, 12, 11],
	"높이" : [50, 40, 20, 50, 20, 30, 40]
})
```

```python
df
```

```python
# sum 합계 계산 axis = 0
df.sum()
```

```python
# sum 합계 계산 axis = 1
df.sum(axis = 1)
```

```python
df.divide(2)
```

```python
df.divide(df.sum(axis = 1), axis = 0)
```

```python
# cumprod 누적곱
df.cumprod(axis = 0)
```

```python
# cumprod 누적 최댓값, 최솟값
df.cummax()
```

```python
df.cummin()
```

```python
df
```

```python
def getVolume(row):
    return row['가로'] * row['세로'] * row['높이']

df['부피'] = df.apply(getVolume, axis = 1)
df
```
